## Validation: Plot the Q-Learning and DQN

### Packages


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Q-Learning

```python
q = pd.read_csv('scores.csv', names=["Count", "Scores"])
q.head()
```

```python
N = 500
x = q['Count']
y = q['Scores']
colors = np.random.rand(N)
area = (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.figure(figsize=(8,5))

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Score achived after 500 iterations', size= 12)
plt.ylabel('Scores')
plt.xlabel('Iterations')
plt.show()
```
